# Hobbies

```python
data2 = {'hobby': [ 'motorcycle riding', 'guitar', 'piano', 'snowboarding', 'fifa', 'chess'], 
                          'amount': [100, 80, 85, 70, 75, 70]}

plt.figure(figsize=(14, 4))
hobbies = sns.barplot(x='hobby', y='amount', data=pd.DataFrame(data2, index=None))
plt.suptitle("My Favorite Hobbies", fontsize=23)
plt.ylabel("Enjoyment Level", fontsize=13)
plt.xlabel("Activity", fontsize=13)
plt.xticks(fontsize=17)
plt.show()

fig = hobbies.get_figure()
fig.savefig('images/hobbies.png')
```
